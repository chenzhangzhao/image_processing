## Color Channel From rgb to grayscale

```python
from skimage import color,data
import matplotlib.pyplot as plt
import numpy as np
import scipy.ndimage as ndi

original = data.coffee()
gray = color.rgb2gray(original)
```

```python
fig,axes = plt.subplots(ncols=2,nrows=1)

axes[0].imshow(gray,cmap='gray')
axes[1].imshow(original)

# what the gray2rgb do is just copy grayscale number to r,g,b channel
# color is an additive system
```

```python
def show_image(image,title='Image',cmap_type='gray'):
    plt.imshow(image,cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

## Numpy for image

```python
# image be represented as numpy array

image = plt.imread('the_Bund.webp')
type(image)
```

```python
# size and shape

print("Size of image: ",image.size)
print("Shape of image:",image.shape)

print(image.shape[0]*image.shape[1]*image.shape[2]==image.size)
```

```python
# color channels
red = image[:,:,0]
green = image[:,:,1]
blue = image[:,:,2]
```

```python
fig,axes=plt.subplots(ncols=4,nrows=1,figsize=(32,8))
axes[0].imshow(image)
axes[1].imshow(red)
axes[2].imshow(green)
axes[3].imshow(blue)
for i in range(4):
    axes[i].axis('off')
```

```python
# image flip 
# Vertically: up and down flipud

ver_img = np.flipud(image)
plt.imshow(ver_img)
plt.show()
```

```python
# Horizontally: left and right fliplr

lr_img = np.fliplr(image)
plt.imshow(lr_img)
plt.show()
```

```python
night_forest = plt.imread('night_forest.jpeg')
snow_forest = plt.imread('snow_forest.jpeg')
```

```python
night_forest = color.rgb2gray(night_forest)
snow_forest = color.rgb2gray(snow_forest)
```

```python
# histogram of different image

fig,axes = plt.subplots(ncols=2,nrows=1,figsize=(24,8))

axes[0].imshow(night_forest,cmap='gray')
axes[0].axis('off')

# in notebook add ; at the end of histplot to suppress the print output of array
axes[1].hist(night_forest.ravel(),bins=256);
```

```python
fig,axes = plt.subplots(ncols=2,nrows=1,figsize=(24,8))

axes[0].imshow(snow_forest,cmap='gray')
axes[0].axis('off')

# in notebook add ; at the end of histplot to suppress the print output of array
axes[1].hist(snow_forest.ravel(),bins=256);
```

```python
hist = ndi.histogram(snow_forest,min=0,max=1,bins=256)
cdf = hist.cumsum()/hist.sum()
plt.plot(cdf)
plt.show()
```
